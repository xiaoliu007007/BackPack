```python
#BP神经网络
%matplotlib qt5
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import Callback
```

```python
from sklearn.preprocessing import scale
```

```python
#训练数据
X = []
Y = []
Z = []
path = '/Users/liuyingjie/Downloads/data/data3/'
filenames = os.listdir(path)
indexs = []
for filename in filenames:
    indexs.append(int(filename))
indexs.sort()
#print(indexs)
for index in indexs:
    file_path = path + str(index)
    with open(file_path,'r',encoding='utf8') as f:
        lines=f.readlines()
    for num in range(34, 54):
        #print(lines[num])
        temps = lines[num].split()
        X.append(float(temps[2])/100)
        #print(float(temps[2])/100)
        Y.append(float(temps[3])/100)
        #print(float(temps[3])/100)
    for line in lines:
        temps = line.split()
        size = len(temps)
        if(size == 11 and temps[0] == '1000000'):
            #print(temps)
            Z.append(float(temps[1])*100000000)
            Z.append(float(temps[6])*100000000)
        if(size == 16 and temps[0] == '1000000'):
            #print(temps)
            Z.append(float(temps[1])*100000000)
            Z.append(float(temps[6])*100000000)
            Z.append(float(temps[11])*100000000)
            
for i in range(len(X)):
    print(Z[i])
            
```

```python
#预测数据
X_pred = []
Y_pred = []
Z_real = []
path = '/Users/liuyingjie/Downloads/data2/'
filenames = os.listdir(path)
indexs = []
for filename in filenames:
    indexs.append(int(filename))
indexs.sort()
for index in indexs:
    file_path = path + str(index)
    with open(file_path,'r',encoding='utf8') as f:
        lines=f.readlines()
    for num in range(34, 54):
        temps = lines[num].split()
        X_pred.append(float(temps[2])/100)
        Y_pred.append(float(temps[3])/100)
    for line in lines:
        temps = line.split()
        size = len(temps)
        if(size == 11 and temps[0] == '1000000'):
            Z_real.append(float(temps[1])*100000000)
            Z_real.append(float(temps[6])*100000000)
        if(size == 16 and temps[0] == '1000000'):
            Z_real.append(float(temps[1])*100000000)
            Z_real.append(float(temps[6])*100000000)
            Z_real.append(float(temps[11])*100000000)
            
for i in range(len(Z_real)):
    print(Z_real[i])
```

```python
X_pred = scale(np.array(X_pred))
Y_pred = scale(np.array(Y_pred))
X_pred_ = np.array([X_pred,Y_pred])
Y_real = scale(np.array(Z_real))
```

```python
from keras.layers import Dense, Activation
#构建神经网络模型
model = Sequential()
model.add(Dense(units = 128, input_dim = 2))
model.add(Activation("relu"))
model.add(Dense(units = 128))
model.add(Activation("relu"))
model.add(Dense(units = 1))
model.add(Activation("linear"))
#构建编译模型
model.compile(loss='mean_squared_error', 
              optimizer='Adam', #Adam
              metrics=['mae'])
model.fit(X_.T, Y_.reshape((-1,1)), epochs = 100, batch_size = 10, shuffle=True, verbose = 1)
```

```python
# 模型评估
loss_and_metrics = model.evaluate(X_.T, Y_, batch_size=100)
loss_and_metrics
```
